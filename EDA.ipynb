## Importamos el dataset

```python
# Una variable para la ruta, buenas prácticas
path_to_data = "./covid_data.csv"

# Importamos el dataset y comprobamos que está correcto
df = pd.read_csv(path_to_data)
# Aumentar el número de líneas a 100
pd.set_option("display.max_columns", 60)
pd.set_option("display.max_rows", 100)
df.head(60)
```

Antes de comenzar el EDA elimino columnas que no voy a estudiar:

```python
df = df.drop(["checkTimeEt","commercialScore","dateChecked","dateModified","fips","grade","hash","negativeRegularScore","negativeScore","positiveScore","score","lastUpdateEt","totalTestResultsSource","pending" ], axis=1)
```

### Transformar las fechas del dataset con un formato legible por python, permitiendo agrupar por mes y año.

```python
#Cambio de formato
df['date'] = pd.to_datetime(df["date"], format="%Y%m%d")
```

```python
#Crear columnas para agrupar por mes y año
df["year"] = df["date"].dt.year
df["month"] = df["date"].dt.month
```

```python
#Agrupamos los valores por mes y año, con una funcion lambda que convierte en lista los valores.
df_agrupado = df.groupby([df["year"], df["month"]]).agg(lambda x: list(x)).reset_index()
```

### Muestra de los grupos por mes y año.

```python
df_agrupado.head(30)
```

### Muestra de Scatter Plots

Se busca mostrar los valores positive, de resultado de test totales y defunciones respecto a la agrupación de los meses.

El objetivo es visualizar datos interesantes como la intermitencia de estos datos.

```python
import matplotlib.pyplot as plt
import numpy as np

# Seleccionamos un subconjunto de columnas para graficar
columnas_a_graficar = ['positive', 'totalTestResults', 'death']

for columna in columnas_a_graficar:
    # Creamos una figura y un conjunto de subtramas
    fig, ax = plt.subplots(figsize=(15, 7))
    
    # Iteramos sobre los grupos
    for i, (name, group) in enumerate(df_agrupado.groupby(["year", "month"])):
        # Convertimos las listas en el dataframe a series de pandas
        series = pd.Series(group[columna].values[0])
        # Creamos una lista de la misma longitud que la serie con la posición actual (i)
        x_values = np.full(series.shape, i)
        # Dibujamos un scatter plot para cada serie
        ax.scatter(x_values, series)
    
    # Configuramos los labels del eje x para que coincidan con los años y meses
    labels = [f'{year}-{month:02d}' for year, month in df_agrupado[["year", "month"]].values]
    ax.set_xticks(range(len(labels)))
    ax.set_xticklabels(labels, rotation=90)
    
    # Configuramos el título y el nombre del eje y
    ax.set_title(f'Scatter plots para {columna}')
    ax.set_ylabel(columna)
    
    # Mostramos el gráfico
    plt.show()
```

### 5. Eliminar variables con una correlación alta. 

```python
print(f'El DataFrame empieza teniendo {df.shape[0]} filas y {df.shape[1]} columnas.')
```

```python
# Calcula la matriz de correlación
corr_matrix = df_mes_especifico.corr().abs()
```

```python
# Selecciona la mitad superior de la matriz de correlación
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
```

```python
# Encuentra los índices de las columnas con correlación mayor que 0.95
to_drop = [column for column in upper.columns if any(upper[column] > 0.95)]
```

```python
# Elimina las columnas altamente correlacionadas
df_mes_especifico.drop(to_drop, axis=1, inplace=True)
```

```python
# Imprime las columnas que van a ser eliminadas
print("Las columnas que van a ser eliminadas debido a la alta correlación son: ", to_drop)
```

```python
# Guarda las columnas originales
original_columns = df_mes_especifico.columns.tolist()
```

```python
# Imprime las columnas que quedan
print("Las columnas que quedan después de eliminar las altamente correlacionadas son: ", df_mes_especifico.columns.tolist())
```

```python
# Imprime las columnas que se han eliminado
deleted_columns = list(set(original_columns) - set(df_mes_especifico.columns.tolist()))
print("Las columnas que se han eliminado son: ", deleted_columns)
```

Los colores más intensos, como el rojo, indican patrones más fuertes, mientras que los colores más claros, como el azul, representan patrones más débiles.

En particular, nos enfocamos en la relación entre las variables "hospitalizedCurrently" y "death". Observamos que existe una correlación muy fuerte entre estas dos variables, como se muestra en el tono rojizo intenso en el gráfico. Además, se asignó un valor de correlación de 0.9, que es muy cercano a 1, lo que indica una correlación muy fuerte entre ambas variables.

Al analizar la relación entre las variables "hospitalizedCurrently" y "death" en nuestro gráfico de calor, se puede observar con bastante claridad una correlación significativa entre ambas. Esto nos indica que, en su momento, las personas que estaban actualmente hospitalizadas tendían a tener una mayor tasa de fallecimiento. Es posible que las personas hospitalizadas presentaran condiciones de salud más graves, lo que podría explicar la mayor incidencia de fallecimientos en ese grupo.

En conclusion, la gravedad de los casos de covid se interelacionan fuertemente como por ejemplo: "hospitalizedCurrently", "death", "ventilatorcurrently".
